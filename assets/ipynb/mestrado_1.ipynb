```python
from pypdf import PdfReader

# Caminho para o arquivo PDF no Google Drive
pdf_path = '/content/drive/MyDrive/Colab Notebooks/dataScience.pdf'

# Inicializa um leitor de PDF
reader = PdfReader(pdf_path)

# Extrai o texto de todas as páginas
pdf_text = ""
for page in reader.pages:
    pdf_text += page.extract_text() + "\n"

print(f'Texto extraído do PDF com sucesso. Total de caracteres: {len(pdf_text)}')
# Opcional: imprimir uma pequena amostra do texto para verificar
# print(pdf_text[:500]) # Imprime os primeiros 500 caracteres
```

```python
# Instala as bibliotecas langchain e langchain-text-splitters, necessárias para dividir o texto em chunks.
!pip install -q langchain langchain-text-splitters

print('Bibliotecas langchain e langchain-text-splitters instaladas com sucesso!')
```

```python
# Importa o Text Splitter da Langchain para dividir o texto em chunks.
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Define o tamanho dos chunks e a sobreposição entre eles
# Um chunk_size de 1000 significa que cada pedaço terá aproximadamente 1000 caracteres.
# Um chunk_overlap de 200 significa que haverá uma sobreposição de 200 caracteres entre os chunks adjacentes.
# Isso ajuda a manter o contexto entre os pedaços de texto.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, # Tamanho máximo de cada chunk
    chunk_overlap=200, # Quantidade de caracteres sobrepostos entre chunks
    length_function=len,
    add_start_index=True,
)

# Divide o texto do PDF em chunks
chunks = text_splitter.create_documents([pdf_text])

print(f'Texto dividido em {len(chunks)} chunks.')
# Opcional: Imprimir o primeiro chunk para ver a estrutura
# print(chunks[0])
```

```python
# Instala as bibliotecas para embeddings (sentence-transformers) e banco de dados vetorial (chromadb).
!pip install -q sentence-transformers chromadb langchain-community

print('Bibliotecas sentence-transformers, chromadb e langchain-community instaladas com sucesso!')
```

```python
# Importa a classe SentenceTransformer para carregar o modelo de embeddings.
from sentence_transformers import SentenceTransformer

# Carrega um modelo de embedding pré-treinado.
# 'all-MiniLM-L6-v2' é um modelo popular por seu bom desempenho e tamanho razoável.
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

print('Modelo de embedding all-MiniLM-L6-v2 carregado com sucesso!')
```

```python
# Importa as classes necessárias do ChromaDB e Langchain para integração.
import chromadb
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings

# Cria uma instância da classe de embeddings que interage com o modelo 'embedding_model' carregado.
# Isso permite que a Langchain utilize o modelo SentenceTransformer para gerar embeddings.
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Cria um banco de dados vetorial Chroma a partir dos chunks e do modelo de embeddings.
# 'persist_directory' define onde o banco de dados será salvo, permitindo que ele seja reutilizado.
# O argumento 'client' é opcional aqui se estiver usando ChromaDB embutido/persistente.
vectorstore = Chroma.from_documents(
    documents=chunks, # Os chunks de texto a serem armazenados.
    embedding=embeddings, # O modelo de embeddings a ser usado.
    persist_directory="./chroma_db" # Diretório onde o banco de dados será salvo.
)

# Persiste (salva) o banco de dados no disco.
vectorstore.persist()

print('Embeddings criados e armazenados no ChromaDB com sucesso!')
print('Banco de dados salvo em: ./chroma_db')
```
